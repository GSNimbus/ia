```python
import pandas as pd

df = pd.read_csv("dataset.csv", sep=";", names=headers_sugeridos, skiprows=9, encoding='UTF-8', decimal=',')
```

```python
df.head()
```

```python
headers_melhores_preditores = [
    "data",
    "hora_utc",
    "precipitacao_total_mm",
    "temperatura_ar_c",
    "umidade_relativa_pct",
    "pressao_atm_estacao_mb",
    "temperatura_orvalho_c",
    "vento_velocidade_ms",

]

df_modelagem  = df[headers_melhores_preditores]

df_modelagem
```

```python
df_modelagem['data_hora'] = pd.to_datetime(df_modelagem['data'] + ' ' + df_modelagem['hora_utc'])
df_modelagem['mes'] = df_modelagem['data_hora'].dt.month
df_modelagem['dia_do_ano'] = df_modelagem['data_hora'].dt.dayofyear
df_modelagem['hora_do_dia'] = df_modelagem['data_hora'].dt.hour
df_modelagem = df_modelagem.drop(['data', 'hora_utc', 'data_hora'], axis=1)
df_modelagem = df_modelagem[df_modelagem['precipitacao_total_mm'] != 0]
```

```python
from sklearn.preprocessing import MinMaxScaler


df_modelagem_copy = df_modelagem.copy().drop(['mes','dia_do_ano','hora_do_dia'],axis=1)

scaler = MinMaxScaler()
df_modelagem_normalizado = scaler.fit_transform(df_modelagem_copy)
df_modelagem_copy = pd.DataFrame(df_modelagem_normalizado, columns=df_modelagem_copy.columns)
df_modelagem_copy = pd.concat([df_modelagem_copy, df_modelagem[['mes','dia_do_ano','hora_do_dia']]], axis=1)
df_modelagem = df_modelagem_copy
```

```python
df_modelagem
```

```python
print(df_modelagem.isnull().sum())
df_modelagem = df_modelagem.dropna()
```

```python
print(df_modelagem.isnull().sum())
```

```python
df_modelagem = df_modelagem[df_modelagem['precipitacao_total_mm'] != 0].copy()
```

```python
df_modelagem
```

```python
from sklearn.model_selection import train_test_split

features = df_modelagem.drop('precipitacao_total_mm', axis=1)
target = df_modelagem['precipitacao_total_mm']
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[len(features.columns)]),
    layers.Dense(64, activation='relu'),
    layers.Dense(1),
])
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001) # Exemplo com Adam
model.compile(loss='mse',
              optimizer=optimizer,
              metrics=['mae', 'mse'])
```

```python
model.summary()
```

```python
history = model.fit(
    x_train, y_train,
    epochs=100,
    verbose=0
)
loss, mae, mse = model.evaluate(x_test, y_test, verbose=2)
```

```python
print(f"Metricas nos dados de teste: MAE={mae:.2f}, MSE={mse:.2f}")

previsoes = model.predict(x_test).flatten()
print("\nAlgumas previsÃµes:")
for i in range(10):
  print(f"Valor real: {y_test.iloc[i]:.2f}, PrevisÃ£o: {previsoes[i]:.2f}")
```
